Get 2 numbers from user, you can assume the first number will be a smaller number while second number is the bigger number. Write a program that will add up the even number from first number to second number. For example if first number is 10 and second number is 20 then the program will add 10,12,14,16,18 and 20 together. Write your program in 2 functions:

1) Function one - Take in 2 numbers as input parameter, add up even numbers within the range of the 2 numbers and return the total

2) Function two - Get the 2 input numbers from the user and call the function one to get the total value and print the output

To get input from user, you may use the input() function and store it into a variable. e.g num = int(input("Enter a first number"))

```python
def calculate_even_sum(start, end):
    total = 0
    for num in range(start, end + 1):
        if num % 2 == 0:
            total += num
    return total

def get_user_input():
    start = int(input("Enter the first number: "))
    end = int(input("Enter the second number: "))
    
    total = calculate_even_sum(start, end)
    print("The sum of even numbers from", start, "to", end, "is:", total)

get_user_input()
```

The goal here is to create a program for a game called "Round 10". In this game, the system randomly generates a number between 1 and 10. It then displays this number and prompts the user to input a number that, when added to the generated number, equals 10. This process continues, with the system generating new numbers and the user responding with the corresponding number needed to reach 10. The game ends when the user correctly responds with the right number three times consecutively.

For instance, if the system generates the number 3, the user should respond with the number 7. If the system generates the number 2, the user should respond with the number 8, and so forth.

You may import the random library and uses its function random.randint(startValue,endValue) to generate the random number.

```python
import random

def game_round():
    consecutive_correct = 0

    while consecutive_correct < 3:
        generated_number = random.randint(1, 10)
        user_answer = int(input(f"Generated number: {generated_number}. Enter a number to add up to 10: "))

        if user_answer + generated_number == 10:
            print("Correct!")
            consecutive_correct += 1
        else:
            print("Incorrect! Try again.")
            consecutive_correct = 0

    print(f"\nCongratulations! You answered correctly 3 times in a row.")

# Run the game
game_round()
```
